```python
z_data.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
z_data.head()
```

```python
z_data.drop_duplicates(inplace=True)
z_data
```

```python
z_data['rate'].unique()
```

```python
def handelerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
z_data['rate']=z_data['rate'].apply(handelerate)
z_data.head()
```

```python
z_data.fillna(z_data['rate'].mean(),inplace=True)
z_data['rate'].isnull().sum()
```

```python
z_data.dropna(inplace=True)
z_data.head()
```

```python
z_data.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
z_data.head()
```

```python
z_data=z_data.drop(['listed_in(city)'],axis=1)
```

```python
z_data['Cost2plates'].unique()
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
z_data['Cost2plates']=z_data['Cost2plates'].apply(handlecomma)
z_data['Cost2plates'].unique()
```

```python
z_data.head()
```

```python
z_data['rest_type'].value_counts()
```

```python
rest_type=z_data['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return'Others'
    else:
        return value
z_data['rest_type']=z_data['rest_type'].apply(handle_rest_type)
z_data['rest_type'].value_counts()
```

```python
cuisines=z_data['cuisines'].value_counts(ascending=False)
cuisines_lessthan1000=cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan1000):
        return 'Others'
    else:
        return value

z_data['cuisines']=z_data['cuisines'].apply(handle_cuisines)
z_data['cuisines'].value_counts()
```
